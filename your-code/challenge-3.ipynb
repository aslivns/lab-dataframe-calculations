## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
customer_info = orders.pivot_table(index= ['CustomerID','Country'], values = ['amount_spent'], aggfunc = {'amount_spent':'sum'}).reset_index()
customer_info.columns = ['CustomerID', 'Country', 'amount_spent']
customer_info.head()
```

```python
upper_limit = np.percentile(customer_info['amount_spent'],95)
lower_limit = np.percentile(customer_info['amount_spent'],75) 
```

```python
VIP_customers = customer_info[(customer_info['amount_spent']>upper_limit)]
VIP_customers = VIP_customers.drop_duplicates()
VIP_customers.head()
```

```python
Preferred_customers = customer_info[(customer_info['amount_spent']>lower_limit) & (customer_info['amount_spent']<upper_limit)]
Preferred_customers.head()
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# UK has the most VIP Customers
VIP_by_country = VIP_customers.groupby(['Country']).count()
VIP_by_country.reset_index()
VIP_by_country.sort_values(by='amount_spent', ascending=False).head(1)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# I found the preferred custermers and concat the tables
# UK has the most
Preferred_by_country = Preferred_customers.groupby(['Country']).count()  
Preferred_by_country.reset_index()
```

```python
VIP_Preferred = pd.concat([VIP_by_country, Preferred_by_country]).groupby(level=0).sum()
VIP_Preferred.sort_values(by='amount_spent', ascending=False).head(1)
```
